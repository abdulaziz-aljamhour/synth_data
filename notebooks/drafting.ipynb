```python
male_names = [
    'محمد', 'علي', 'سلمان', 'فهد', 'عبدالله', 'حسين', 'يوسف', 'سامي', 'ماجد', 'عادل',
    'عبدالعزيز', 'تركي', 'عبدالرحمن', 'صالح', 'خالد', 'مبارك', 'خلف', 'بدر', 'فيصل', 'إبراهيم'
]

female_names = [
    'فاطمة', 'نور', 'مريم', 'خديجة', 'عائشة', 'مها', 'نورة', 'سارة', 'هدى', 'عبير',
    'أمل', 'شهد', 'حنان', 'رنا', 'سلمى', 'شذى', 'رجاء', 'زينب', 'رغد', 'أسماء'
]


family_names = [
    'الزهراني', 'التميمي', 'الغامدي', 'الشمري', 'العتيبي', 'الحربي', 'القحطاني', 'الراشد',
    'المالكي', 'الحربي', 'الفيصل', 'الشيباني', 'القصيبي', 'البقمي', 'البلوي', 'العصيمي',
    'الحسيني', 'الفهيد', 'السبيعي', 'الداوود'
]

# Convert them to DataFrames
df_male_names = pd.DataFrame(male_names, columns=["FIRST_NAME"])
df_female_names = pd.DataFrame(female_names, columns=["FIRST_NAME"])
df_family_names = pd.DataFrame(family_names, columns=["FAMILY_NAME"])

def generate_phone_number():
    return "05" + ''.join(random.choices(string.digits, k=8))


def generate_national_id():
    return ''.join(random.choices(string.digits, k=10))


# Function to generate a random birth date between 1980 and 2010
def generate_birth_date():
    year = random.randint(1980, 2010)
    month = random.randint(1, 12)
    day = random.randint(1, 28)  # To avoid issues with days in months
    return datetime(year, month, day)

# Function to generate a random username
def generate_username(user_id):
    return f"user{user_id}"

# Function to generate a random password
def generate_password(user_id):
    return f"password{user_id}"

# Function to generate random user data
def generate_user_data(user_id):
    first_name = random.choice(df_male_names["FIRST_NAME"].tolist() + df_female_names["FIRST_NAME"].tolist())
    middle_name = random.choice(df_male_names["FIRST_NAME"].tolist())
    family_name = random.choice(df_family_names["FAMILY_NAME"].tolist())
    phone_number = generate_phone_number()
    national_id = generate_national_id()
    birth_date = generate_birth_date()
    region_id = random.randint(1, 13)
    username = generate_username(user_id)
    password = generate_password(user_id)
    creation_date = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    
    return {
        "USER_ID": user_id,
        "NATIONAL_ID": national_id,
        "FIRST_NAME": first_name,
        "MIDDLE_NAME": middle_name,
        "FAMILY_NAME": family_name,
        "PHONE_NUMBER": phone_number,
        "BIRTH_DATE": birth_date,
        "REGION_ID": region_id,
        "USER_NAME": username,
        "PASSWORD": password,
        "CREATION_DATE": creation_date
    }

# Create a list of users
users = [generate_user_data(i) for i in range(1, 1001)]

# Convert the list of users to a pandas DataFrame
df_users = pd.DataFrame(users)

df_users.head()
```

```python
conn = lite.connect(ROOT + '/data/database/ecommerce.db')
df_users.to_sql('USERS', conn, if_exists='append', index=False)
conn.commit()
conn.close()
```

```python
import sqlite3

# Connect to the SQLite database
conn = lite.connect(ROOT + '/data/database/ecommerce.db')
cursor = conn.cursor()

# Query to fetch all table names from sqlite_master
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

# Loop through each table and count rows
row_counts = {}
for table in tables:
    table_name = table[0]
    try:
        # Dynamically count rows in each table
        cursor.execute(f"SELECT COUNT(*) FROM {table_name}")
        row_count = cursor.fetchone()[0]
        row_counts[table_name] = row_count
    except sqlite3.Error as e:
        # Handle potential errors, such as tables with special constraints
        print(f"Error counting rows in table {table_name}: {e}")

# Print the row counts for each table
for table_name, row_count in row_counts.items():
    print(f"Table: {table_name}, Rows: {row_count}")

# Close the connection
conn.close()
```

```python
# Function to generate the shipping contract agreement
def generate_shipping_contract_agreement(contract_id, seller_ids, shipper_ids, service_level_ids):
    seller_id = random.choice(seller_ids)
    shipper_id = random.choice(shipper_ids)
    contract_duration = random.randint(1, 24)  # Contract duration in months (1 to 24)
    status = random.choice([0, 1])  # 0: inactive, 1: active
    service_level_id = random.choice(service_level_ids)
    begins_at = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    ends_at = (datetime.now().replace(year=datetime.now().year + 1)).strftime('%Y-%m-%d %H:%M:%S')
    created_at = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    updated_at = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    
    return {
        "SHIPPING_CONTRACT_ID": contract_id,
        "SELLER_ID": seller_id,
        "SHIPPER_ID": shipper_id,
        "CONTRACT_DURATION": contract_duration,
        "STATUS": status,
        "SERVICE_LEVEL_ID": service_level_id,
        "BEGINS_AT": begins_at,
        "ENDS_AT": ends_at,
        "CREATED_AT": created_at,
        "UPDATED_AT": updated_at
    }

# Define shippers, sellers, and service levels
seller_ids = list(range(1, 20))  # 19 sellers
shipper_ids = list(range(1, 11))  # 10 shippers
service_level_ids = list(range(1, 6))  # 5 service levels

# Generate a list of 50 shipping contract agreements
contracts = [generate_shipping_contract_agreement(i, seller_ids, shipper_ids, service_level_ids) for i in range(1, 51)]

# Convert the list to a pandas DataFrame
df_shipping_contracts = pd.DataFrame(contracts)

# Display the first few rows
df_shipping_contracts.head()

conn = lite.connect(ROOT + '/data/database/ecommerce.db')
df_shipping_contracts.to_sql('SHIPPING_CONTRACT_AGREEMENT', conn, if_exists='append', index=False)
conn.commit()
conn.close()
```

```python
import string
# Function to generate random orders
def generate_order_data(order_id, seller_ids, user_ids, product_ids, shipper_ids):
    order_status_id = random.randint(1, 4)  # Assuming 4 order statuses
    payment_status_id = random.randint(1, 3)  # Assuming 3 payment statuses
    product_id = random.choice(product_ids)
    user_id = random.choice(user_ids)
    seller_id = random.choice(seller_ids)
    subtotal = random.randint(50, 1000)  # Random subtotal between 50 and 1000
    shipping_cost = random.randint(10, 100)  # Random shipping cost between 10 and 100
    total = subtotal + shipping_cost  # Total = subtotal + shipping cost
    payment_method_id = random.randint(1, 4)  # Assuming 4 payment methods
    tracking_number = ''.join(random.choices(string.digits, k=10))  # Random 10-digit tracking number
    notes = random.choice(["", "Please deliver after 5 PM", "Fragile", "Gift wrap this order"])
    created_at = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    updated_at = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    ordered_at = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    shipped_at = datetime.now().strftime('%Y-%m-%d %H:%M:%S') if order_status_id in [2, 3] else None  # Only shipped if status is 'Shipping' or 'Completed'
    
    return {
        "ORDER_ID": order_id,
        "ORDER_STATUS_ID": order_status_id,
        "PAYMENT_STATUS_ID": payment_status_id,
        "PRODUCT_ID": product_id,
        "USER_ID": user_id,
        "SELLER_ID": seller_id,
        "TOTAL": total,
        "SUBTOTAL": subtotal,
        "SHIPPING_COST": shipping_cost,
        "PAYMENT_METHOD_ID": payment_method_id,
        "TRACKING_NUMBER": tracking_number,
        "NOTES": notes,
        "CREATED_AT": created_at,
        "UPDATED_AT": updated_at,
        "ORDERED_AT": ordered_at,
        "SHIPPED_AT": shipped_at
    }

# Define seller_ids, user_ids, product_ids, and shipper_ids based on the previously generated data
seller_ids = list(range(1, 20))  # 19 sellers
user_ids = list(range(1, 1001))  # 1000 users
product_ids = list(range(1, 1001))  # 1000 products
shipper_ids = list(range(1, 11))  # 10 shippers

# Generate 1000 orders
orders = [generate_order_data(i, seller_ids, user_ids, product_ids, shipper_ids) for i in range(1, 1001)]

# Convert the list to a pandas DataFrame
df_orders = pd.DataFrame(orders)

# Display the first few rows
df_orders.head()
conn = lite.connect(ROOT + '/data/database/ecommerce.db')
df_orders.to_sql('ORDERS', conn, if_exists='append', index=False)
conn.commit()
conn.close()
```
